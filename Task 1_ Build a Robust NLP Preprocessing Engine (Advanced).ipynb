# Task 1: Conceptual Understanding



## 1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" can be treated as two different words because of case sensitivity. Computers distinguish between uppercase and lowercase letters. If we do not normalize the text, both may be counted separately, leading to duplication. That is why converting text to lowercase is an important preprocessing step.

---

## 2. What happens if stopwords are not removed?

If stopwords are not removed, the text will contain many frequently occurring words like "is", "the", and "and", which do not contribute much to the actual meaning. This increases noise in the data and may affect model performance, as important words may get less importance.

---

## 3. Provide two real-world scenarios where removing stopwords can be harmful.

One example is sentiment analysis. Words like "not" are often treated as stopwords, but removing them can change the meaning. For example, "not good" becomes "good", which gives the wrong interpretation.

Another example is question-answering systems. Words like "who", "what", and "where" help identify the type of question. Removing them may make it difficult for the system to understand the user's intent.

---

## 4. Explain the difference between stemming and lemmatization.

Stemming reduces words to their root form by removing prefixes or suffixes. It is a fast process but may produce incorrect words. For example, "running" may become "runn".

Lemmatization converts words to their base or dictionary form using context and meaning. For example, "running" becomes "run". It is more accurate than stemming but slightly slower.

# Task 2:Build Advanced Preprocessing Function





```python
import re

def preprocess_text(text):

    # 1. lowercase
    text = text.lower()

    # 2. remove urls
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # 4. remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. handle repeated characters (FINAL WORKING)
    text = re.sub(r'(.)\1+', r'\1\1', text)   # keep max 2 letters
    text = re.sub(r'\bso+\b', 'so', text)     # fix "sooo" → "so"

    # 6. remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 7. remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. remove short words
    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['i', 'is', 'no', 'not', 'so']]

    text = " ".join(words)

    return text
```

```python
texts = [
    "I have 2 dogs!!!",
    "This is good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now"
]

for t in texts:
    print(preprocess_text(t))
```

# Task 4: Token Analytics


```python
def token_analysis(text):
    tokens = text.split()
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    if total_tokens == 0:
        avg_length = 0
    else:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    
    return total_tokens, unique_tokens, avg_length


# Use same cleaned texts
for t in texts:
    cleaned = preprocess_text(t)
    
    total, unique, avg = token_analysis(cleaned)
    
    print("Sentence :", cleaned)
    print("Total tokens :", total)
    print("Unique tokens :", unique)
    print("Average token length :", round(avg, 2))
    print()
```

# Task 5: Frequency Analysis


```python
from collections import Counter

# Step 1: Collect all tokens
all_tokens = []

for t in texts:
    cleaned = preprocess_text(t)
    tokens = cleaned.split()
    all_tokens.extend(tokens)

# Step 2: Count frequency
word_freq = Counter(all_tokens)

# Step 3: Top 10 most frequent words
top_10 = word_freq.most_common(10)

# Step 4: Top 5 least frequent words
least_5 = sorted(word_freq.items(), key=lambda x: x[1])[:5]

# Output
print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word} : {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word} : {count}")
```

# Task 6: Build Full Pipeline


def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        cleaned = preprocess_text(text)
        tokens = cleaned.split()
        
        clean_sentences.append(cleaned)
        all_tokens.extend(tokens)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

```python
result = full_pipeline(texts)

print("Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```

```python
def preprocess_text(text):
    
    # Handle None or empty input
    if not text or not isinstance(text, str):
        return ""
    
    # your existing code...
    import re
    
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    words = text.split()
    words = [w for w in words if len(w) > 1 or w in ['no', 'not']]
    
    text = " ".join(words)
    
    return text
```
