# Chatbot using Transformers (Hugging Face)

## Objective
Build a simple chatbot using a pre-trained transformer model that can interact with users and generate meaningful responses.

## Pipeline
User Input → Model Processing → Response Generation → Output → Loop until Exit

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

## Model Loading
Loading a pre-trained transformer model (DialoGPT) from Hugging Face.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "microsoft/DialoGPT-medium"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

## Chatbot Function
This function takes user input and generates a response using the transformer model.

```python
import torch

def generate_response(user_input, chat_history_ids):
    
    # Encode input
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    # Append history
    if chat_history_ids is not None:
        input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
    else:
        input_ids = new_input_ids

    # Generate response
    chat_history_ids = model.generate(
        input_ids,
        max_length=200,
        pad_token_id=tokenizer.eos_token_id,
        attention_mask=torch.ones(input_ids.shape, dtype=torch.long),
        do_sample=True,
        top_k=40,
        top_p=0.9,
        temperature=0.5
    )

    # Decode response
    response = tokenizer.decode(chat_history_ids[:, input_ids.shape[-1]:][0], skip_special_tokens=True)

    # Filter empty responses
    if len(response.strip()) < 2:
        response = "I'm here to help! Please ask something."

    return response, chat_history_ids
```

## Chatbot Interaction
The chatbot continuously takes input and responds until the user types 'exit' or 'quit'.

```python
print("Chatbot: Hello! I am your AI assistant. How can I help you today?")


chat_history_ids = None

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: You're welcome! Have a nice day.")
        break

    elif user_input.lower() in ["hello", "hi"]:
        print("Chatbot: Hello! Nice to meet you. How can I assist you today?")
        continue

    elif "thank you" in user_input.lower():
        print("Chatbot: You're welcome! Feel free to ask more questions.")
        continue

    # Prompt improvement
    if "artificial intelligence" in user_input.lower():
        user_input = "What is Artificial Intelligence? Give a clear and correct definition."

    elif "who created python" in user_input.lower():
        user_input = "Who created Python programming language? Give correct answer."

    # Transformer response
    response, chat_history_ids = generate_response(user_input, chat_history_ids)
    print("Chatbot:", response)
```
